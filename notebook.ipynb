```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
top_hosts = df['host_id'].value_counts().head(10).index
top_hosts_df = df[df['host_id'].isin(top_hosts)].copy()
```

```python
top_hosts_df['host_label'] = top_hosts_df['host_id'].astype(str) + ' - ' + top_hosts_df['host_name']
host_ratings = top_hosts_df.groupby('host_label')['rating'].mean().sort_values(ascending= False)
```

```python
# Top 10 Hosts by Listings and their Average Ratings
host_ratings.plot(kind='bar', title='Top 10 Hosts by Listings and their Average Ratings', color='green')
plt.ylabel("Average Rating")
plt.xlabel("Host ID - Host Name")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Exclude rows with no host_name or rating
df_filtered = df.dropna(subset=['host_name', 'rating'])

# Get hosts with at least 3 listings (to avoid outliers with a single bad review)
host_counts = df_filtered['host_id'].value_counts()
eligible_hosts = host_counts[host_counts >= 3].index

# Filter and create combined host label
low_rating_df = df_filtered[df_filtered['host_id'].isin(eligible_hosts)].copy()
low_rating_df['host_label'] = low_rating_df['host_id'].astype(str) + ' - ' + low_rating_df['host_name']

# Group by host label and calculate average rating
low_rating_avg = low_rating_df.groupby('host_label')['rating'].mean().sort_values().head(10)
```

```python
# Top 10 Hosts With The Lowest Average Ratings (≥3 listings)
low_rating_avg.plot(kind='bar', title='Top 10 Hosts With The Lowest Average Ratings', color='red')
plt.ylabel("Average Rating")
plt.xlabel("Host ID - Host Name")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Geographical Distribution of Airbnb Listings in New York City and Classification by Neighbourhood Groups
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='neighbourhood_group')
plt.title("Geographical Distribution of Airbnb Listings in NYC and Classification by Neighbourhood Groups")
plt.show()
```

```python
# Geographical Distribution of Different Types of Rooms Available in New York City

g = sns.FacetGrid(df, col='room_type', col_wrap=3, height=5) # Set up the FacetGrid

# Map the scatterplot onto the grid
g.map_dataframe(sns.scatterplot, x='longitude', y='latitude', alpha=0.6)

g.set_titles(col_template="{col_name}")
g.set_axis_labels("Longitude", "Latitude")
g.figure.suptitle("Geographical Distribution of Different Types of Rooms Available in New York City", fontsize=16)
g.figure.tight_layout()
g.figure.subplots_adjust(top=0.9)
plt.show()
```

```python

# Geographical Distribution of Different Types of Rooms Available in Different Neighbourhood Groups
room_palette = {
    'Entire home/apt': 'orange',
    'Private room': 'blue',
    'Shared room': 'green',
    'Hotel room': 'red'
}
g = sns.FacetGrid(
    df,
    col='neighbourhood_group',
    col_wrap=3,
    height=5,
    hue='room_type',
    palette=room_palette,
    sharex=True,
    sharey=True
)
g.map_dataframe(sns.scatterplot, x='longitude', y='latitude', alpha=0.6, s=20)
g.set_titles("{col_name}")
g.set_axis_labels("Longitude", "Latitude")
g.add_legend(title="Room Type", loc='lower right')
g.figure.suptitle("Geographical Distribution of Different Types of Rooms Available in Different Neighbourhood Groups", fontsize=16)
g.figure.tight_layout()
g.figure.subplots_adjust(top=0.92)
plt.show()
```
